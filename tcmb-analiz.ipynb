**Verilerimizi temizledik analizimizi yapalım.**
**Zamana göre işsizlik oranının değişimine bakalım.**

```python
plt.figure(figsize=(12,6))
plt.plot(data["Tarih"],data["issizlik_orani"],label="İşsizlik Oranları")
plt.title("Tarihe Göre İşsizlik Oranları")
plt.xlabel("İşsizlik Oranları")
plt.ylabel("Tarih")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
data_new = data.select_dtypes(include=["number"])
corr_matrix = data_new.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True,cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```
